```python
# Changing the date to day only
airport_df['day_of_week'] = airport_df['Date'].dt.day_name()
```

```python
#Lets Change the Dtype of Some Columns
airport_df = airport_df.astype({'Date':str})
```

```python
# split column and add new columns to Data frame
airport_df[['Origin', 'Destination']] = airport_df['Route'].str.split('-', expand=True)
```

```python
# Lets find distance between two sources of flights
airport_df['Distance'] = ((airport_df['Geometry Coordinates 0 0'] - airport_df['Geometry Coordinates 1 0'])**2 + (airport_df['Geometry Coordinates 1 1'] - airport_df['Geometry Coordinates 0 1'])**2)
airport_df['Distance'] = ((airport_df['Distance'])**0.5)*110
```

```python
airport_df['City'] = np.where(airport_df['Origin'] != 'SFO', airport_df['Origin'], airport_df['Destination'])
```

```python
#Lets Change the Dtype of Some Columns
airport_df = airport_df.astype({'Date':str})

# Creating plot between Date and number of Flights 
airport_df['Date'].value_counts().sort_values().plot.bar()
plt.rcParams['figure.figsize'] = (12,6)
plt.title('Date Wise Flight Count',fontsize=25)
plt.xlabel('Date', fontsize=15)
plt.xticks (rotation= 90) 
plt.ylabel('Number of Flights', fontsize=15)
plt.show()
```

```python
# Creating plot between Date and number of Flights 
airport_df['day_of_week'].value_counts().sort_values().plot.bar()
plt.rcParams['figure.figsize'] = (12,6)
plt.title('Day Wise Flight Count',fontsize=25)
plt.xlabel('Date', fontsize=15)
plt.xticks (rotation= 90) 
plt.ylabel('Number of Flights', fontsize=15)
plt.show()
```

```python
# Creating plot between Date and number of Flights 
airport_df['Properties Flysfo Gate'].value_counts().sort_values().plot.bar()
plt.rcParams['figure.figsize'] = (24,6)
plt.title('Flight With Gate Count',fontsize=25)
plt.xlabel('Date', fontsize=15)
plt.xticks (rotation= 90) 
plt.ylabel('Number of Flights', fontsize=15)
plt.show()
```

```python
# Finding percentage of churn in regular customers
airport_df['Properties Flysfo Event'].value_counts().plot.pie(explode=[0.05,0.1],autopct='%1.1f%%',startangle=180,shadow=True,figsize=(9,9))
plt.title("Departure And Arrival Rate",size = 18)
plt.show()
```

```python
# Identifying the Arrival and Departure Flight
Arrival_data = airport_df[airport_df['Properties Flysfo Event'] == 'A']
Departure_data = airport_df[airport_df['Properties Flysfo Event'] == 'D']
```

```python
Arrival_data.to_csv("output.csv")
```

```python
# Bar Plot of No. Of Arrival Flights vs Place  
Arrival_data['City'].value_counts().sort_values().tail(10).plot.bar()
plt.rcParams['figure.figsize'] = (12,6)
plt.title('Most Flight Arrival To',fontsize=25)
plt.xlabel('Destination', fontsize=15)
plt.xticks (rotation= 0) 
plt.ylabel('Number of Flights', fontsize=15)
plt.show()
```

```python
# Bar Plot of No. Of Departure Flights vs Place  
Departure_data['City'].value_counts().sort_values().tail(10).plot.bar()
plt.rcParams['figure.figsize'] = (12,6)
plt.title('Most Flight Departure From',fontsize=25)
plt.xlabel('Destination', fontsize=15)
plt.xticks (rotation= 0) 
plt.ylabel('Number of Flights', fontsize=15)
plt.show()
```

```python
# Most Distances From SFO
airport_df.groupby(['City'])['Distance'].mean().sort_values().tail(10).plot.bar()
plt.rcParams['figure.figsize'] = (12,6)
plt.title('Most Distances From SFO',fontsize=25)
plt.ylabel('Destination', fontsize=15)
plt.xticks (rotation= 0) 
plt.xlabel('Distance', fontsize=15)
plt.show()
```

```python
# Lets go a bit deeper. Analyzing churn with respect to active international plan
flight = airport_df.loc[airport_df['Properties Flysfo Base Airline'] != 'UA']
flight ['Properties Flysfo Base Airline'].value_counts().sort_values().tail(10).plot.pie(autopct ='%1.1f%%',shadow = True)
plt.title("Different Base Airline Count",size = 20)
plt.show()
```

```python
# Lets go a bit deeper. Analyzing churn with respect to active international plan
flight = airport_df.loc[airport_df['Properties Flysfo Airline'] != 'UA']
flight ['Properties Flysfo Airline'].value_counts().sort_values().tail(10).plot.pie(autopct ='%1.1f%%',shadow = True)
plt.title("Different Airline Count",size = 20)
plt.show()
```

```python
# Delay of flights
airport_df['delay'] = (airport_df['Properties Flysfo Estimated Timestamp'] - airport_df['Properties Flysfo Actual Timestamp'])/60 #divide by 60 for output in minute
airport_df['delay'].astype({'delay':'int'}) #changing the datatype to int
df1 = airport_df[['delay','City']]
df1.groupby(['City'])['delay'].mean().sort_values().plot.bar()
plt.rcParams['figure.figsize'] = (25,6)
plt.title('Delay of Flights',fontsize=25)
plt.ylabel('Delay in Minutes', fontsize=15) 
plt.xlabel('Destinations', fontsize=15)
plt.show()
```

```python
df1 = airport_df[['delay','Date']]
df1.groupby(['Date'])['delay'].mean().sort_values().plot.bar(x= 'Origin' , y = 'delay')
plt.rcParams['figure.figsize'] = (25,6)
plt.title('Delay of Flights',fontsize=25)
plt.ylabel('Delay in Minutes', fontsize=15) 
plt.xlabel('Date', fontsize=15)
plt.show()
```
